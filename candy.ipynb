```python
import tabletools as tt

t = tt.read_csv('candy-data.csv')

#display the last 4 rows of the data set
print(t.tail(4))

#show the candy name, if it's chocolate, if it's peanuty/almondy, and its win percentage for all candy that is chocolate
choco_t = tt.Table([line for line in t.data if line[1] == 1], columns=t.columns)

narrow_choco_t = choco_t[['competitorname', 'chocolate', 'peanutyalmondy', 'winpercent']]

print(narrow_choco_t)


#using the results above, add another criteria: only candies that are peanuty/almondy 
#(that is… at the end, you'll only have a listing of chocolate and peanut candies!)
choco_and_pa_t = tt.Table([line for line in narrow_choco_t.data if line[2] == 1], columns=narrow_choco_t.columns)

print(choco_and_pa_t)

#using the previous results, apply each of the following filters on chocolate and almond-peanut candies, with one "filter" per cell

#chocolate and almond-peanut candies that have win percentage < than 50
## first-lambda
t_with_filter1 = tt.Table(
    list(filter(lambda x: x[3] < 50, choco_and_pa_t.data)),
    columns=choco_and_pa_t.columns
)

print(t_with_filter1)

#chocolate and almond-peanut candies with a name that starts with "Reese", YOU MUST USE A LAMBDA FOR THIS
## second-lambda
t_with_filter2 = tt.Table(
    list(filter(lambda x: x[0].startswith('Reese'), choco_and_pa_t.data)),
    columns=choco_and_pa_t.columns
)

print(t_with_filter2)

#chocolate and almond-peanut candies with a name less than 10 characters long. YOU MUST USE A LAMBDA FOR THIS
## third-lambda
t_with_filter3 = tt.Table(
    list(filter(lambda x: len(x[0]) < 10, choco_and_pa_t.data)),
    columns=choco_and_pa_t.columns
)

print(t_with_filter3)
```
